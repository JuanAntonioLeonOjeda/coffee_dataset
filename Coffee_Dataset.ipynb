### Imports

Here we import our dependencies

```python
import os
import pandas as pd
import zipfile
import urllib.request
```

### Import Dataset

Download dataset

```python
zip_url = "https://www.kaggle.com/api/v1/datasets/download/fatihb/coffee-quality-data-cqi"
full_route = os.path.join("res", "dataset.zip")
urllib.request.urlretrieve(zip_url, full_route)
```

Unzip file

```python
file_route = os.path.join("res")
with zipfile.ZipFile(full_route, 'r') as zip:
    zip.extractall(file_route)
```

```python
dataset_url = os.path.join("res", "df_arabica_clean.csv")

coffee_dataset = pd.read_csv(dataset_url, sep=",")

display(coffee_dataset)
```
